```python
import pandas
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
```

```python
PATH = '/Users/pujandave/cider/tests/test_results_05_08_21_01_08'
df = pandas.read_csv(f'{PATH}/test.csv')
fig, ax = plt.subplots(figsize=(20,6))
styles = ['-','--','-']
count = 0
for name, group in df.groupby('compute_node'):
    print(f'Average Effective Load on Node {name} : {group.time.value_counts().mean()}')
    group.time.value_counts().sort_index().plot(ax=ax, label=name, style=styles[count])
    count+=1
    count = count%3
    
plt.xlabel("Time (s)")
plt.ylabel("Effective Load")
plt.title('Capability Based Node Selection')
plt.linestyle = styles
plt.legend(title='Compute Nodes')
plt.savefig("SimpleNodeSelection.pdf")
```
