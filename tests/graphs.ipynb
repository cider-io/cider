```python
import pandas
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('/Users/pujandave/Big Folder/UIUC Acedemics/Grad School/CS-525/Project/Tests/test-1-1000-100/test.csv')

fig, ax = plt.subplots(figsize=(20,6))

for name, group in df.groupby('request_node'):
    group.plot(x='iteration', y='total_duration', ax=ax, label=name)
    
plt.xlabel("Iteration")
plt.ylabel("Total Duration (s)")
plt.title('All to Randomly Picked VM #10 with data of 1000 floats (100 iterations)')
plt.legend(title='Request Nodes')
plt.savefig("test-1-1000-100-total-duration.pdf")

fig, ax = plt.subplots(figsize=(20,6))

for name, group in df.groupby('request_node'):
    group.plot(x='iteration', y='compute_time', ax=ax, label=name)
    
plt.xlabel("Iteration")
plt.ylabel("Compute Time (s)")
plt.title('All to Randomly Picked VM #10 with data of 1000 floats (100 iterations)')

plt.legend(title='Request Nodes')

plt.savefig("test-1-1000-100-compute-time.pdf")
```
